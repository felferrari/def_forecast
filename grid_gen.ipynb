```python
import geopandas as gpd
import pandas as pd
```

```python
cells_shp = r'data/ibama_shp/cells.shp'
deter_shp = r'data/deter/deter-amz-deter-public.shp'
```

```python
cells_gpd = gpd.read_file(cells_shp)
deter_gpd = gpd.read_file(deter_shp)
```

```python
cells_gpd['CELL_AREA'] = cells_gpd['geometry'].area
```

```python
deter_gpd['VIEW_DATE'] = pd.to_datetime(deter_gpd['VIEW_DATE'])
deter_gpd
```

```python
deter_period = deter_gpd.loc[(deter_gpd['VIEW_DATE']>= '2020-01-01') & (deter_gpd['VIEW_DATE'] <= '2020-01-16')]
```

```python
merged = gpd.overlay(deter_period.to_crs('EPSG:4326'), cells_gpd, how = 'intersection')
merged['INTER_AREA'] = merged['geometry'].area
merged
```
